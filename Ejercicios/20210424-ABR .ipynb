```r
datos<-Surv(rats$time, rats$status) 
datos
```

```r
KM.datos<-survfit(datos~1)
KM.datos
```

```r
summary(KM.datos)
```

```r
plot(KM.datos, main="Supervivencia estimada")
```

```r
t<-c(20, 70, 120)
```

```r
km.hm<-survfit(datos~rats$sex)
km.hm
```

```r
survdiff(datos~rats$sex, rho=0)
```
